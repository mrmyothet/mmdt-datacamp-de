These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend.

```python
best_math_schools = schools[["school_name", "average_math"]].sort_values(by="average_math", ascending=False)
best_math_schools.head()
```

```python
schools["total_SAT"] = schools.apply(lambda row: row["average_math"] + row["average_reading"] + row["average_writing"], axis=1)
top_10_schools = schools[["school_name", "total_SAT"]].sort_values(by="total_SAT", ascending=False)
top_10_schools
```

```python
average_SAT = schools.groupby("borough")["total_SAT"].agg(["min", "max", "median"]).round(2)
average_SAT
```
